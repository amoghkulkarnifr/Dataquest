```

plt.show()
```

# FiveThirtyEight style

<div><p>Previously, we calculated the correlation values between wine quality and wine properties. We want to represent these values visually in the form of two bar plots. Below, you can see what we want to build.</p>
<p><img src="https://s3.amazonaws.com/dq-content/528/screen1_1.png" alt="img"></p>
<p>Let's start by creating the two bar plots using the <code>fivethirtyeight</code> style.</p>
</div>

```
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(9, 5))
ax.barh(white_corr.index, white_corr)
ax.barh(red_corr.index, red_corr)
plt.show()
```

<div>
<p><img src="https://s3.amazonaws.com/dq-content/528/screen3_1.png" alt="img"></p>
<p>We can tell from the two colors that Matplotlib built two different bar plots. It's not clear which is which, though.</p>
<p>To separate them, we're going to move the white wine plot to the right side of the graph. To do that, we can use the <code>left</code> parameter of the <code>Axes.barh()</code> method.</p>
<p>This parameter moves the left sides of the bars from their default x-coordinate of <code>0</code>. In our case, <code>left=2</code> moves them from <code>0</code> to <code>2</code> (for bars representing negative quantities, the right sides move from <code>0</code> to <code>2</code>).</p>
</div>

```
# Assume the other code is written
ax.barh(white_corr.index, white_corr, left=2)
ax.barh(red_corr.index, red_corr)
```

<div>
<p><img src="https://s3.amazonaws.com/dq-content/528/screen3_2.png" alt="img"></p>
<p>Based on the principle of proximity, we now see two different bar plots: the red wine plot on the left (red-colored), and the white wine plot on the right (blue-colored).</p>
<p>To maximize the data-ink ratio, let's now erase the following:</p>
<ul>
<li>Non-data ink</li>
<li>Redundant data ink</li>
</ul>
<p>Although you don't see the code, note that these have remained saved from the previous screen: <code>red_wine</code>, <code>white_wine</code>, <code>red_corr</code>, <code>white_corr</code>.</p></div>

### Instructions 

<ol>
<li>Remove the grid of the graph using the <a href="https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.grid.html" target="_blank"><code>Axes.grid()</code> method</a>. Read the documentation to find the parameter you need to use.</li>
<li>Remove the x- and y-tick labels. Use the <a href="https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_xticklabels.html" target="_blank"><code>Axes.set_xticklabels()</code> method</a> and the <a href="https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_yticklabels.html" target="_blank"><code>Axes.set_yticklabels()</code> method</a>.</li>
<li>Reduce the width of each bar to <code>0.5</code> by using the <code>Axes.barh()</code> method — you'll need to use the <code>height</code> parameter.</li>

````python
# Initial Code

style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(9, 5))
ax.barh(white_corr.index, white_corr, height=0.5, left=2)
ax.barh(red_corr.index, red_corr, height=0.5)

ax.grid(b=False)
ax.set_xticklabels([])
ax.set_yticklabels([])

plt.show()
````

# Adding a signature

<div><p>FiveThirtyEight graphs generally have a signature bar at the bottom, where we see the creator's name and the data source.</p>
<p><img src="https://s3.amazonaws.com/dq-content/528/m4_1.png" alt="img"></p>
<p>Matplotlib doesn't provide us with a function dedicated to generating signature bars. Again, we need to be creative and use the existing tools in creative ways.</p>
<p>One way to draw a signature bar is to add text with a dark background color. Below, we add the text <code>©DATAQUEST     Source: P. Cortez et al.</code> with a background color of <code>#4d4d4d</code> (a shade of grey). The text itself has the color <code>#f0f0f0</code>, which is the same as the graph's background color.</p>
</div>

```
# Assume the rest of the code is written
ax.text(-0.7, -2.9,
        '©DATAQUEST     Source: P. Cortez et al.',
        color = '#f0f0f0',
        backgroundcolor = '#4d4d4d',
        size=12)
```

<div>
<p><img src="https://s3.amazonaws.com/dq-content/528/screen6_1.png" alt="img"></p>
<p>To create the visual effect of a signature, we need to add more white space between <code>©DATAQUEST</code> and <code>Source: P. Cortez et al.</code>. We can use multiplication to add multiple white spaces — <code>' '*94</code> adds 94 white spaces.</p>

# Coloring bars differently 

<div><p>The last change we're going to make to our graph is the bar colors:</p>
<ul>
<li>Positive correlation values will take the color <code>#33A1C9</code> (a tint of blue).</li>
<li>Negative correlation values will take the color <code>#ffae42</code> (a tint of orange).</li>
</ul>
<p><img src="https://s3.amazonaws.com/dq-content/528/m4_2.png" alt="img"></p>
<p>To make this change, we'll use the <code>color</code> parameter of <code>Axes.barh()</code>. So far, we've only used strings with this parameter: <code>'blue'</code>, <code>'red'</code>, <code>'#af0b1e'</code>, etc.</p>
<p>The <code>color</code> parameter, however, can also take in an array of strings, where each string represents a color. Below, we use a Python <code>list</code> to specify the colors for four bars — note that Matplotlib preserves the order of colors in the list:</p>
</div>

```
colors = ['red', 'orange', 'purple', 'green']
plt.bar(['a', 'b', 'c', 'd'], [2, 4, 7, 3],
        color=colors)
plt.show()
```

</div>
