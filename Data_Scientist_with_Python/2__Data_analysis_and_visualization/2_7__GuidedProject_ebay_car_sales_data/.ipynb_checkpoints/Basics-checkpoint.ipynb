# 2. Cleaning column names 

```python
autos.columns
```

```python
corrected_column_names = {
    "dateCrawled" : "ad_crawled",
    "offerType" : "offer_type",
    "vehicleType" : "vehicle_type",
    "yearOfRegistration": "registration_year",
    "powerPS" : "power_ps",
    "monthOfRegistration" : "registration_month",
    "fuelType" : "fuel_type",
    "notRepairedDamage" : "unrepaired_damage",
    "nrOfPictures" : "n_pictures",
    "postalCode" : "postal_code",
    "lastSeen" : "last_seen",
    "dateCreated" : "ad_created"
}

new_column_names = []

for c in autos.columns:
    if c in corrected_column_names:
        new_column_names.append(corrected_column_names[c])
    else:
        new_column_names.append(c)

autos.columns = new_column_names
```

```python
autos.columns
```

* Notes
  - Changed column names from `camelCase` to `snake_case` for consistency and readability
  - some one worded column names are unchanged

# 3. Initial exploration and data cleaning 

```python
autos.describe()
```

* Notes
  - `n_pictures` column contains all zeroes, can be dropped
  - `registration-year` column has minimum `1000` which is dubious
  - `power_ps` column has minimum `0` which is dubious
  - `postal_code` has some 4-digit values which needs to be made `null`
  - All of them can be converted to `int` from `float64`

```python
autos["price"] = autos["price"].str.replace("$", "")
autos["price"] = autos["price"].str.replace(",", "")

autos["price"] = autos["price"].astype(float)
```

```python
autos["odometer"] = autos["odometer"].str.replace("km", "")
autos["odometer"] = autos["odometer"].str.replace(",", "")

autos["odometer"] = autos["odometer"].astype(int)
```

```python
# Rename odometer 
autos.rename({"odometer" : "odometer_km"}, axis=1, inplace=True)
```

# 6. Dealing with incorrect `registration_year` data

```python
# Finding out what are the highest and lowest 
# acceptable values for the `registration_year`

autos["valid_registration_year"] = \
    (autos["registration_year"] <= autos["ad_created"].str[:4].astype(int)) \
    & (autos["registration_year"] >= 1910) \
    & (autos["registration_year"] <= 2016)
```

```python
autos["valid_registration_year"].value_counts()
```

* Notes 
  - A lot of entries in the table are such that the `registration_year` value is greater than the `ad_created` value, which means when the ad was created, the vehicle was not registered. 
  - Also for the sanity check, the entries where `registration_year` is less than `1910` and greater than `2016` are omitted.

# 7. Exploring `price` by `brand` 

```python
top_10_brands_by_n_posts = autos["brand"].value_counts(normalize=True, dropna=False).sort_values(ascending=False).head(10)
print(top_10_brands_by_n_posts)
```

```python
shortlisted_cars_2 = autos[
    (autos["valid_registration_year"] == True) &
    (autos["registration_year"].between(2013, 2016)) &
    (autos["odometer_km"] < 30000)
]
shortlisted_cars_2.describe()

mean_prices_by_brand = {}

for b in top_10_brands_by_n_posts.index:
    shortlisted_b = shortlisted_cars_2[shortlisted_cars_2["brand"] == b]
    mean_prices_by_brand[b] = shortlisted_b["price"].describe()["mean"]
```

```python
mean_prices_by_brand    # In our shortlisted cars 
```

```python
# Which brands are more likely to be put for sale after the car has 
# run >1,00,000 km 

shortlisted_cars_3 = autos[
    (autos["valid_registration_year"] == True) &
    (autos["odometer_km"] > 100000)
]

resold_brands = shortlisted_cars_3["brand"].value_counts()
total_cars_brands = autos["brand"].value_counts()

# Brands that are more likely to be put to resale after running more than 100000 
more_likely_to_be_resold = resold_brands / total_cars_brands
more_likely_to_be_resold.sort_values(ascending=False).head(15)
```

* Notes
  - `more_likely_to_be_resold_` can also be interpreted as "the brands that people are more likely to hold on to till they run beyond 100000 miles". The top performers in this series are the ones that people tend to hold on to until they have run more than 100000 miles. 
  - `mean_prices_by_brand` are the mean prices of the brands that are most present in the database. 

# 8. Storing aggregate data in a DataFrame

```python
mean_mileage_by_brands = {}
mean_price_by_brands = {}

for b in autos["brand"].unique():
    mean_mileage_by_brands[b] = autos[autos["brand"] == b]["odometer_km"] \
                                    .describe()["mean"]

for b in autos["brand"].unique():
    mean_price_by_brands[b] = autos[autos["brand"] == b]["price"] \
                                    .describe()["mean"]
```
