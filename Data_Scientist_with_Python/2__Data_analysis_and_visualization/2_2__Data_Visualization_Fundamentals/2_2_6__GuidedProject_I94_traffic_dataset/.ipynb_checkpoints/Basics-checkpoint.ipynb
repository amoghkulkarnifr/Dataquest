# Traffic volume: Day vs Night

```python
i94_traffic['date_time'] = pd.to_datetime(i94_traffic['date_time'])
```

```python
datetimes = i94_traffic['date_time']
daytime_entries = i94_traffic[(datetimes.dt.hour >= 7) & (datetimes.dt.hour < 19)]
nighttime_entries = i94_traffic[(datetimes.dt.hour < 7) | (datetimes.dt.hour >= 19)]
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
daytime_entries['traffic_volume'].plot.hist(bins=1000)
plt.ylim(0, 250)
plt.title('Daytime traffic volume')

plt.subplot(1, 2, 2)
nighttime_entries['traffic_volume'].plot.hist(bins=1000)
plt.ylim(0, 250)
plt.title('Nighttime traffic volume')

plt.show()
```

```python
daytime_entries.describe()
```

```python
nighttime_entries.describe()
```

* Notes 
  - The plots have very distinct profiles, one being left skewed and the other one being right skewed. 
  - Daytime plot shows mostly normal curve, with mean being at 4500 - meaning that at significant number of hours during the day, the traffic volume was in the range of 3000/hr to 6000/hr.
  - Nighttime plot shows mostly exponentially decreasing curve, with a peak at 500 - meaning that the most number of hours the traffic was between 0 to 1000/hr.

# Time indicators

```python
# Create a month column
daytime_entries['month'] = daytime_entries['date_time'].dt.month
# Group by newly created column's values 
daytime_monthwise = daytime_entries.groupby('month').mean()

nighttime_entries['month'] = nighttime_entries['date_time'].dt.month
nighttime_monthwise = nighttime_entries.groupby('month').mean()
```

```python
months = daytime_entries['month'].unique()

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
daytime_monthwise['traffic_volume'].plot.line()
plt.title('Avg. Daytime traffic volume by month')
plt.xticks(months)
plt.ylim(0, 5500)

plt.subplot(1, 2, 2)
nighttime_monthwise['traffic_volume'].plot.line()
plt.title('Avg. Nighttime traffic volume by month')
plt.xticks(months)
plt.ylim(0, 5500)

plt.show()
```

```python
daytime_monthwise.describe()
```

```python
nighttime_monthwise.describe()
```

* Notes 
  - Both line graphs show decline towards 1st and 12th month, as these are the months of peak winter. 
  - Month of July has a strange dip on both graphs, but it is more pronounced in the daytime graph.
  - Nighttime graph is significantly lower than the daytime graph, which reinforces the prior observation about the over traffic volume during day vs. night.

# Time indicators (II)

```python
# Create a day-of-the-week column
daytime_entries['dow'] = daytime_entries['date_time'].dt.dayofweek
# Group by newly created column's values 
daytime_daywise = daytime_entries.groupby('dow').mean()

nighttime_entries['dow'] = nighttime_entries['date_time'].dt.dayofweek
nighttime_daywise = nighttime_entries.groupby('dow').mean()
```

```python
days = daytime_entries['dow'].unique()

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
daytime_daywise['traffic_volume'].plot.line()
plt.title('Avg. Daytime traffic volume by DoW')
plt.xticks(days)
plt.ylim(0, 5500)

plt.subplot(1, 2, 2)
nighttime_daywise['traffic_volume'].plot.line()
plt.title('Avg. Nighttime traffic volume by DoW')
plt.xticks(days)
plt.ylim(0, 5500)

plt.show()
```

* Notes
  - As expected, traffic volume plummets on Saturdays and Sundays for both daytime and nighttime and then comes slowly back up throughout the rest of the days of a week.
  - Reduction during the daytime is more pronounced than that of the nighttime.

# Time indicators (III)

```python
daytime_entries['hour'] = daytime_entries['date_time'].dt.hour
bussiness_days = daytime_entries.copy()[daytime_entries['dow'] <= 4] # 4 = Friday
weekend = daytime_entries.copy()[daytime_entries['dow'] > 4] # 5 = Saturday, 6 = Sunday

by_hour_business = bussiness_days.groupby('hour').mean()
by_hour_weekend = weekend.groupby('hour').mean()
```

```python
hours = daytime_entries['hour'].unique()

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
by_hour_business['traffic_volume'].plot.line()
plt.title('Avg. Daytime traffic volume on weekdays (by hour)')
plt.xticks(hours)
plt.ylim(0, 7000)

plt.subplot(1, 2, 2)
by_hour_weekend['traffic_volume'].plot.line()
plt.title('Avg. Daytime traffic volume on weekends (by hour)')
plt.xticks(hours)
plt.ylim(0, 7000)

plt.show()
```

* Notes
  - Rush hours at business days and slack hours on weekends are clearly visible in these line plots.
  - Highest mean traffic on the weekends is almost as much as the lowest mean traffic on the business days.

### Summary of observations 

1. The traffic is as much as twice during the daytime of that of the nighttime.
2. The traffic is more during the summer months, both during the daytime and the nighttime.
3. On weekends the traffic is very low as compared to the business days. 
4. The rush hours during the day are in the early morning and in the evening (4pm-5pm) on the weekdays. 

**Hence, if you want to get bored to death in traffic for a long time, get your car out by 7AM or by 4pm on a Wednesday in the month of July.**

# Weather indicators

```python
i94_traffic_weather.corr()
```

```python
i94_traffic_weather.plot.scatter(x='temp', y='traffic_volume')
plt.xlim(225, 350)
plt.xlabel('Temperature')
plt.ylabel('Traffic volume')
plt.show()
```

* Notes
  - The scatter plot shows very slight correlation between temperature and traffic volume.
  - The correlations between other weather related columns with the traffic volume are even weaker than that of the temperature.
