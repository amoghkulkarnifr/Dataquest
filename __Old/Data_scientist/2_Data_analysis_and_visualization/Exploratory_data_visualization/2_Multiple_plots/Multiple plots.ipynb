```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])

xvalues = unrate.loc[0:11]['DATE']
yvalues = unrate.loc[0:11]['VALUE']

plt.plot(xvalues, yvalues)
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')

plt.show()
```

# Matplotlib Classes

- Use `plt.figure()` to create a figure and assign to `fig`.
- Use `Figure.add_subplot()` to create two subplots above and below each other
- Assign the top `Axes` object to `ax1`.
- Assign the bottom `Axes` object to `ax2`.
- Use `plt.show()` to display the resulting plot.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
plt.show()
```

# Adding Data

Create 2 line subplots in a 2 row by 1 column layout:
- In the top subplot, plot the data from 1948.
- In the bottom subplot, plot the data from 1949.

```python
%matplotlib inline

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

xvalues_1 = xvalues
yvalues_1 = yvalues
ax1.plot(xvalues_1, yvalues_1)

xvalues_2 = unrate.loc[12:23]['DATE']
yvalues_2 = unrate.loc[12:23]['VALUE']
ax2.plot(xvalues_2, yvalues_2)
plt.show()
```

# Formatting and Spacing

For the plot we generated in the last screen, set the width of the plotting area to 12 inches and the height to 5 inches.

```python
%matplotlib inline

fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

xvalues_1 = xvalues
yvalues_1 = yvalues
ax1.plot(xvalues_1, yvalues_1)

xvalues_2 = unrate.loc[12:23]['DATE']
yvalues_2 = unrate.loc[12:23]['VALUE']
ax2.plot(xvalues_2, yvalues_2)



plt.show()
```

# Comparing Across More Years

- Set the width of the plotting area to `12` inches and the height to `12` inches.
- Generate a grid with 5 rows and 1 column and plot data from the individual years. Start with 1948 in the top subplot and end with 1952 in the bottom subplot.
- Use `plt.show()` to display the plots.

```python
fig = plt.figure(figsize=(12, 12))

for i in range(5):
    subplot = fig.add_subplot(5, 1, i+1)
    
    start = 0+12*i
    end = 11+12*i
    rows = unrate.loc[start:end]
    xvalues = rows['DATE']
    yvalues = rows['VALUE']

    subplot.plot(xvalues, yvalues)

plt.show()
```

# Overlaying Line Charts

- Set the plotting area to a width of `6` inches and a height of `3` inches.
- Generate 2 line charts in the base subplot, using the `MONTH` column for the x-axis instead of the `DATE` column

```python
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(6, 3))

plt.plot(unrate.loc[0:11]['MONTH'], unrate.loc[0:11]['VALUE'], c='red')
plt.plot(unrate.loc[12:23]['MONTH'], unrate.loc[12:23]['VALUE'], c='blue')

plt.show()
```

Let's do the same for all the five months.

```python
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(10, 6))

plt.plot(unrate.loc[0:11]['MONTH'], unrate.loc[0:11]['VALUE'], c='red')
plt.plot(unrate.loc[12:23]['MONTH'], unrate.loc[12:23]['VALUE'], c='blue')
plt.plot(unrate.loc[24:35]['MONTH'], unrate.loc[24:35]['VALUE'], c='green')
plt.plot(unrate.loc[36:47]['MONTH'], unrate.loc[36:47]['VALUE'], c='orange')
plt.plot(unrate.loc[48:59]['MONTH'], unrate.loc[48:59]['VALUE'], c='black')

plt.show()
```

# Adding a legend

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
year = 1948
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=str(year))
    year += 1
plt.legend(loc='upper left')
plt.show()
```
